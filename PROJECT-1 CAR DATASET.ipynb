```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("car.csv")
```

```python
data
```

```python
results=data[["selling_price","km_driven"]].describe()
results["selling_price"]=results["selling_price"].astype("int")
results
```

```python
data.index=data["name"]
```

```python
data['Company']=data['name'].apply(lambda x:x.split(" ")[0])
data
```

```python
results=data["Company"].value_counts()
results
```

```python
data["fuel"].value_counts()
```

```python
data[data['fuel']=="Electric"]
```

```python

results
```

```python
results.plot.bar()
plt.show()
```

```python
data[data['Company']=="Jeep"]
```

```python
results=data.groupby("Company")[["selling_price"]].mean()
results["selling_price"]=results["selling_price"].astype("int")
results.sort_values(by=["selling_price"])
```

```python
results.plot.bar()
plt.show()
```

```python
results=data.groupby("Company")[["km_driven"]].mean()
results.sort_values(by=["km_driven"])
results.plot.bar()
plt.show()
```

```python
data["owner"].unique()
```

```python
owner_count = data.groupby("owner")["owner"].count()
owner_count.plot.pie(autopct='%.2f%%', title="Pie chart") 
plt.show()
```

```python
data["transmission"].unique()
```

```python
transmission_count = data.groupby("transmission")["transmission"].count()
transmission_count.plot.pie(autopct='%.2f%%', title="Pie chart") 
plt.show()
```
